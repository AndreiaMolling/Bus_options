```python
def Menu():
    
    menu = input('Escolha a opção: 1) Escolher viagem, 2) Cancelar reserva, 3) Encontrar assentos, 4) Dados da rota, 5) Sair')
    
    if menu == '1':
        Onibus()
        print('Concluído')
    elif menu == '2':
        Cancelar_reserva()
    elif menu == '3':
        Encontrar_Assentos()
    elif menu == '4':
        Dados()
    elif menu == '5':
        exit()
    
    else:
        print('error')
        BV()
    return 
```

```python
def Onibus():
    destino = input('Escolha seu destino: 1)Porto Alegre - Florianopolis, 2) Criciuma-Florianopolis, 3)Porto Alegre - Criciuma')
    if destino == '1':
        P_F()
    elif destino == '2':
        C_F()
    elif destino == '3':
        P_C()
        
    else:
        print('error')
        BV()
    return
```

```python
def Encontrar_Assentos():
    reserva = input('Qual o destino:1)Porto Alegre - Florianopolis, 2) Criciuma-Florianopolis, 3)Porto Alegre - Criciuma')
    if reserva == '1':
        o = input('Horário: 1) 6h, 2) 16h')
        if o == '1':
            EA('R', assento6)
        elif o == '2':
            EA('R', assento16)
        else:
            print('error')
            BV()
    elif reserva == '3':
        o = input('Horário: 1) 6h, 2) 16h')
        if o == '1':
            EA('R2',assento6)
        elif o == '2':
            EA('R2',assento16)
        else:
            print('error')
            BV()
    elif reserva == '2':
        o = input('Horário: 1) 10h, 2) 20h')
        if o == '1':
            EA('R3', assento6_t)
        elif o == '2':
            EA('R3', assento16_t)
        else:
            print('error')
            BV()
    else:
        print('error')
        BV()
```

```python
def EA(marker, assento ):
    display(assento)
    e=[]
    num = int(input('Número de assentos'))
    for i in range(len(assento)):
        if (assento[i] == ' '):
            e.append(i)
    j=0
    c =0
    t=[]
    for j in range(len( e)):
        if len(t) == num:
            print(t)
            break
        else:
            t = list(set(t))
            if len(t) == num:
                print(t)
                break
            else:
                if e[j] == max(e):
                    break
        
                elif (e[j] < e[j+1]) and (e[j+1] - e[j] == 1):
                    c = c +1
                    t.append(e[j])
                    t.append(e[j+1])
                    t = list(set(t))
        
                else:
                    c = 0
                    t=[]
                    continue
    
    if len(t) == num:
        print('Há',num, 'assentos consecutivos:', t)
        for k in t:
            assento[k] = marker
        print('Reservado')
        display(assento)
        Menu()
    else:
        print('Não há',num,'assentos consecutivos')
        Menu()
        

        
```
